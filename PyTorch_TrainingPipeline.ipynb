## Model Building begins

```python
# Yaha pe we take OOPs wala appraoch.

class MySimpleNN():

  def __init__(self,X):
    self.weights = torch.rand(size=(X.shape[1],1),dtype=torch.float64,requires_grad=True)
    self.bias = torch.zeros((1),requires_grad=True,dtype=torch.float64)

  def forward(self,X):
    z = torch.matmul(X,self.weights) + self.bias
    return torch.sigmoid(z)

  def loss_function(self,y_pred,y):

    epsilon = 1e-7
    y_pred = torch.clamp(y_pred,epsilon,1-epsilon)

    loss = -(y*torch.log(y_pred) + (1-y)*torch.log(1-y_pred)).mean()
    return loss


```

```python
# Defining Learning rate and epochs
learning_rate = 0.1
epochs = 25
```

```python
## Model Pipeline

model = MySimpleNN(X_train_tensor)

for epoch in range(epochs):
  y_pred = model.forward(X_train_tensor)
  #print(y_pred)

  loss = model.loss_function(y_pred,y_train_tensor)
  #print(f"Epoch {epoch+1},Loss {loss.item()}")

  loss.backward()

  with torch.no_grad():
    model.weights -= learning_rate*model.weights.grad
    model.bias -= learning_rate*model.bias.grad

  model.weights.grad.zero_()
  model.bias.grad.zero_()

  print(f"Epoch: {epoch+1},Loss : {loss.item()}")
```

```python
## Evaluation

with torch.no_grad():
  y_pred = model.forward(X_train_tensor)
  y_pred = (y_pred>0.5).float()

print(y_pred)


```
